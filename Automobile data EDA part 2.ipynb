```python
import numpy as np 
import pandas as pd
from pandas import DataFrame
import seaborn as sns
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot


```

```python
df = pd.read_csv('automobile.csv')
df
```

```python
df.replace('?',np.nan, inplace = True )
df.head()
#replace ? with NaN
```

```python
#store all the nan values where false is present and true is missing
missing_values = df.isnull()
missing_values.head(5)
```

```python
for column in df.columns.values.tolist():
    print(column)
    print(df[column].value_counts())
    print('')
```

```python
df.columns
```

```python
# Finding mean of a specific column by making a def
average_hp=df["horsepower"].astype('float').mean(axis=0)
print("AVERAGE HORSE POWER IS:",average_hp)
 #we can repeat same for peak rpm, normalized losses price etc
```

```python
#now replace this average value for all nan values
df['horsepower'].replace(np.nan, average_hp, inplace = True)
df['horsepower'].isnull
```

```python
average_bore=df["bore"].astype('float').mean(axis=0)
print("AVERAGE BORE IS:",average_bore)
df['bore'].replace(np.nan, average_bore, inplace = True)
df['bore'].isnull
```

```python
average_norl=df["normalized-losses"].astype('float').mean(axis=0)
print("AVERAGE NORL IS:",average_norl)
df['normalized-losses'].replace(np.nan, average_norl, inplace = True)
df['normalized-losses'].isnull
```

```python
#no null value should be in price range
df.dropna(subset=['price'],axis=0,inplace=True)
df.head()
```

```python
df.dtypes
```

```python
#converting datatypes
df['normalized-losses']=df['normalized-losses'].astype('int')
df[['bore', 'price']]=df[['bore', 'price']].astype('float')
df[['peak-rpm', 'stroke']]=df[['peak-rpm', 'stroke']].astype('float')
```

```python
#normalizing
df['width']=df['width']/df['width'].max()
df['length']=df['length']/df['length'].max()
```

```python
#binning
df['horsepower']=df['horsepower'].astype(int,copy = True)
bins = np.linspace(min(df['horsepower']) , max(df['horsepower']),4)
intensity = ['High','Medium','Low']
df['horse-power-binned'] = pd.cut(df['horsepower'],bins,labels=intensity,include_lowest=True)
df[['horsepower', 'horse-power-binned']].head(10)


```

```python
df['horse-power-binned'].value_counts()
```

```python
#visualizing
pyplot.bar(intensity,df["horse-power-binned"].value_counts())
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title("horse-power-binned")
```

```python
#categorical values to numerical
#Dummies
var1 = pd.get_dummies(df['fuel-type'])
var1.head()
var1.rename(columns={'gas_fuel':'gas','diesel_fuel':'diesel'}, inplace =True)
var1.head()
```

```python
#merge these data frames into one
df = pd.concat([df,var1],axis=1)
```

```python
df.columns
```

```python
#removing the fuel-type column
df.drop("fuel-type", axis=1, inplace=True)
```

```python
df.to_csv('df_wrangled.csv')
```
